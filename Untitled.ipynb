```python
sales_per_product.rename(columns = {'price': 'total_sales'}, inplace = True)
```

```python
products = pd.merge(products, sales_per_product, on = 'product_id')
```

```python
products.sort_values('total_sales', ascending = False)
```

```python
products_dummy = pd.get_dummies(products, columns=['product_category_name'])
```

```python
products_dummy.dropna(inplace = True)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
X = products_dummy.drop(columns = ['total_sales', 'product_id'])
y = products_dummy['total_sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
lr = LinearRegression()
```

```python
ss = StandardScaler()
```

```python
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
X_train_sc
```

```python
lr.fit(X_train_sc, y_train)
```

```python
lr.score(X_train_sc, y_train)
```

```python
lr.score(X_test_sc, y_test)
```
